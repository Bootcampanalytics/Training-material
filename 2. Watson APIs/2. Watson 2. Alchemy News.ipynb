```python
import requests
import json
def print_api_resp(url_parameters,credentials):
    response = requests.get(credentials_1["url"]+"/data/GetNews?"+url_parameters+"&apikey="+credentials_1['apikey'])  
    parsed = json.loads(response.text)
    print json.dumps(parsed, indent=4, sort_keys=True)
```

### Basic example

```python

url_parameters="outputMode=json&start=now-1d&end=now&maxResults=3&return=enriched.url.title&q.enriched.url.title=data%20science"
print_api_resp(url_parameters,credentials_1)
```

### Dedup example

```python
url_parameters="outputMode=json&start=now-1d&end=now&count=3&q.enriched.url.title=president&return=enriched.url.cleanedTitle&dedup=1"
print_api_resp(url_parameters,credentials_1)
```

### Rank example

```python
url_parameters="rank=high^medium&start=now-1d&end=now&return=enriched.url.url&outputMode=json&count=2"
print_api_resp(url_parameters,credentials_1)
```

### Relations example

This example returns the 5 most recent articles that mention a company acquiring another company the article title. This parameter specifies that we want to search for a relation in the article title where subject entity type is Company, object entity type is Company and verb linking the subject and object is 'acquire'.

```python
url_parameters="outputMode=json&start=now-1d&end=now&count=5&q.enriched.url.enrichedTitle.relations.relation=|subject.entities.entity.type=Company,action.verb.text=acquire,object.entities.entity.type=Company|&return=enriched.url.url,enriched.url.title"
print_api_resp(url_parameters,credentials_1)
```

### Entity example

This example returns the 5 most recent articles that mention Apple the company (q.enriched.url.enrichedTitle.entities.entity.text), and not apple the fruit (q.enriched.url.enrichedTitle.entities.entity.type. It also specifies to return article titles and urls (return=enriched.url.url,enriched.url.title)

```python
url_parameters="outputMode=json&start=now-1d&end=now&count=5&q.enriched.url.enrichedTitle.entities.entity=|text=apple,type=company|&return=enriched.url.url,enriched.url.title"
print_api_resp(url_parameters,credentials_1)
```

### Taxonomy example

This example returns the 5 most recent articles about movies (q.enriched.url.enrichedTitle.taxonomy.taxonomy_.label=movies) with article URLs and titles (return=enriched.url.url,enriched.url.title.

```python
url_parameters="outputMode=json&start=now-1d&end=now&count=5&q.enriched.url.enrichedTitle.taxonomy.taxonomy_.label=movies&return=enriched.url.url,enriched.url.title"
print_api_resp(url_parameters,credentials_1)
```

### Sentiment analysis example
This example returns the 5 most recent article titles (q.enriched.url.title=IBM) with positive sentiment about IBM. q.enriched.url.enrichedTitle.docSentiment=|type=positive,score=>0.5 specifies that the sentiment should be positive and greater than 0.5.

```python
url_parameters="outputMode=json&start=now-1d&end=now&count=5&q.enriched.url.title=IBM&q.enriched.url.enrichedTitle.docSentiment=|type=positive,score=>0.5|&return=enriched.url.url,enriched.url.title"
print_api_resp(url_parameters,credentials_1)
```

### Concept example

This example returns the 5 most recent articles tagged with the concept football (q.enriched.url.concepts.concept.text=football).

```python
url_parameters="outputMode=json&start=now-1d&end=now&count=5&q.enriched.url.concepts.concept.text=football&return=enriched.url.url,enriched.url.title"
print_api_resp(url_parameters,credentials_1)
```

### Keywords example

This example returns the 5 most recent articles that have the keyword paintings in their URL and title (q.enriched.url.enrichedTitle.keywords.keyword.text=paintings).

```python
url_parameters="outputMode=json&start=now-1d&end=now&count=5&q.enriched.url.enrichedTitle.keywords.keyword.text=paintings&return=enriched.url.url,enriched.url.title"
print_api_resp(url_parameters,credentials_1)
```

# Making calls with the Watson library

```python
import json
from watson_developer_cloud import AlchemyDataNewsV1

alchemy_data_news = AlchemyDataNewsV1(api_key=credentials_1['apikey'])

results = alchemy_data_news.get_news_documents(start='now-7d', end='now',
                                               time_slice='12h')
print(json.dumps(results, indent=2))

results = alchemy_data_news.get_news_documents(
    start='1453334400',
    end='1454022000',
    return_fields=['enriched.url.title',
                   'enriched.url.url',
                   'enriched.url.author',
                   'enriched.url.publicationDate'],
    query_fields={
        'q.enriched.url.enrichedTitle.entities.entity':
            '|text=IBM,type=company|'})
print(json.dumps(results, indent=2))
```
